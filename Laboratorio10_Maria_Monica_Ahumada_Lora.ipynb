## **SECCIÓN 1: INTERVALOS DE CONFIANZA PARA DIFERENCIA DE MEDIAS (MUESTRAS GRANDES)**
Contexto 1: Brecha salarial por género en el sector financiero
Un estudio del Ministerio del Trabajo analiza la posible brecha salarial entre hombres y mujeres en el sector financiero colombiano. Se tomaron muestras independientes de 150 hombres y 150 mujeres. Los resultados fueron:

Género	n	Media salarial (millones COP)	Desviación estándar
Hombres	150	$4.8	$1.2
Mujeres	150	$4.2	$1.1
Pregunta 1. ¿Cuál es el intervalo de confianza del 95% para la verdadera diferencia de medias salariales (μH - μM)?

* (0.32, 0.88)
* (0.45, 0.75)
* (0.28, 0.92)
* (0.38, 0.82)

Pregunta 2. Con base en el intervalo anterior, ¿qué conclusión es correcta?

* No hay evidencia suficiente para afirmar que existe brecha salarial
* La brecha salarial es estadísticamente significativa al 95% de confianza
* Las mujeres ganan más que los hombres en el sector financiero
* El intervalo no es válido porque las varianzas son diferentes







```python
# =============================================================================
# @title PYTHON SECCIÓN 1: DIFERENCIA DE MEDIAS – MUESTRAS GRANDES (Z)
# =============================================================================
print("=" * 65)
print("  SECCIÓN 1: INTERVALOS DE CONFIANZA – MUESTRAS GRANDES")
print("=" * 65)

# ── Contexto 1: Brecha salarial ───────────────────────────────────────────────
n_h, x_h, s_h = 150, 4.8, 1.2
n_m, x_m, s_m = 150, 4.2, 1.1
diff1 = x_h - x_m
se1   = np.sqrt(s_h**2/n_h + s_m**2/n_m)
z95   = 1.960
li1_p1 = diff1 - z95 * se1
ls1_p1 = diff1 + z95 * se1

print(f"\n📊 PREGUNTA 1 – IC 95% brecha salarial (μH − μM)")
print(f"   IC 95%: ({li1_p1:.2f} , {ls1_p1:.2f})")
print(f"   ✅ Respuesta: A")

print(f"\n📊 PREGUNTA 2 – Conclusión")
print(f"   El intervalo NO contiene el 0 → diferencia significativa.")
print(f"   ✅ Respuesta: B")

# ── Contexto 2: Programa social ───────────────────────────────────────────────
n_b, x_b, s_b = 200, 850, 180
n_nb, x_nb, s_nb = 200, 780, 175
diff2 = x_b - x_nb
se2   = np.sqrt(s_b**2/n_b + s_nb**2/n_nb)
z99   = 2.576
margen2 = z99 * se2
li2 = diff2 - margen2
ls2 = diff2 + margen2

print(f"\n📊 PREGUNTA 3 – IC 99% programa social (μB − μNB)")
print(f"   IC 99%: ({li2:.1f} , {ls2:.1f})")
print(f"   ✅ Respuesta: B")

print(f"\n📊 PREGUNTA 4 – Conclusión")
print(f"   El IC está completamente por encima de 0.")
print(f"   ✅ Respuesta: B")

# ─────────────────────────────────────────────────────────────────────────────
# 📊 GRÁFICA SECCIÓN 1
# ─────────────────────────────────────────────────────────────────────────────
fig1, axes1 = plt.subplots(2, 1, figsize=(12, 7))
fig1.suptitle(f'Sección 1 – IC para Diferencia de Medias (Muestras Grandes)\n{NOMBRE}',
              fontsize=13, fontweight='bold', color='#1A237E')

plot_ic_horizontal(axes1[0], li1_p1, ls1_p1, diff1, cero=0,
                   titulo='P1-P2 | IC 95%: Brecha Salarial (μH − μM) [millones COP]',
                   etiqueta=f'IC 95%: ({li1_p1:.2f}, {ls1_p1:.2f})')

plot_ic_horizontal(axes1[1], li2, ls2, diff2, cero=0,
                   titulo='P3-P4 | IC 99%: Impacto Programa Social (μB − μNB) [miles COP]',
                   etiqueta=f'IC 99%: ({li2:.1f}, {ls2:.1f})', color='#9C27B0')

plt.tight_layout()
plt.savefig('IC_Seccion1_MariaAhumadaLora.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Gráfica guardada: IC_Seccion1_MariaAhumadaLora.png")
```

Contexto 3: Comparación de metodologías educativas

1.4.2 📊 Contexto 4: Efecto de una campaña de salud pública
Una universidad evalúa el efecto de una campaña de alimentación saludable en dos barrios de estrato socioeconómico similar. Barrio A (intervención) n=15, Barrio B (control) n=13.

Barrio	n	Consumo de frutas (porciones/semana)	s

Pregunta 10. Si se asume varianzas diferentes (prueba de Welch), los grados de libertad aproximados serían:

* 24.5
* 25.8
* 26.2
* 23.9

```python
# =============================================================================
# @title PYTHON SECCIÓN 2: DIFERENCIA DE MEDIAS – MUESTRAS PEQUEÑAS (t de Student)
# =============================================================================
print("\n" + "=" * 65)
print("  SECCIÓN 2: INTERVALOS DE CONFIANZA – MUESTRAS PEQUEÑAS")
print("=" * 65)

# ── Contexto 3: Metodologías educativas ──────────────────────────────────────
n_t, x_t, s_t = 12, 3.8, 0.6
n_i, x_i, s_i = 10, 4.3, 0.5
diff3 = x_i - x_t
gl3 = n_t + n_i - 2
sp2_3 = ((n_t - 1)*s_t**2 + (n_i - 1)*s_i**2) / gl3
sp_3  = np.sqrt(sp2_3)
t90_20 = stats.t.ppf(0.95, df=gl3)
se3 = sp_3 * np.sqrt(1/n_i + 1/n_t)
margen3 = t90_20 * se3
li3 = diff3 - margen3
ls3 = diff3 + margen3

print(f"\n📊 PREGUNTA 5 – Valor crítico t")
print(f"   t(0.05, {gl3}) = {t90_20:.3f} → ✅ Respuesta: A")

print(f"\n📊 PREGUNTA 6 – IC 90% (μI − μT)")
print(f"   IC 90%: ({li3:.2f} , {ls3:.2f}) → ✅ Respuesta: C")

print(f"\n📊 PREGUNTA 7 – Varianza combinada")
print(f"   Sp² = {sp2_3:.4f} → ✅ Respuesta: B")

# ── Contexto 4: Campaña de salud pública ─────────────────────────────────────
n_a, x_a, s_a = 15, 12.5, 2.8
n_b2, x_b2, s_b2 = 13, 10.2, 3.0
diff4 = x_a - x_b2
gl4_igual = n_a + n_b2 - 2
sp2_4 = ((n_a-1)*s_a**2 + (n_b2-1)*s_b2**2) / gl4_igual
sp_4  = np.sqrt(sp2_4)
t95_26 = stats.t.ppf(0.975, df=gl4_igual)
se4   = sp_4 * np.sqrt(1/n_a + 1/n_b2)
li4   = diff4 - t95_26*se4
ls4   = diff4 + t95_26*se4

# Welch df
num_w = (s_a**2/n_a + s_b2**2/n_b2)**2
den_w = (s_a**2/n_a)**2/(n_a-1) + (s_b2**2/n_b2)**2/(n_b2-1)
gl4_welch = num_w / den_w

print(f"\n📊 PREGUNTA 8 – GL (varianzas iguales): {gl4_igual} → ✅ A")
print(f"\n📊 PREGUNTA 9 – Interpretación: IC no contiene 0 → ✅ C")
print(f"\n📊 PREGUNTA 10 – GL Welch: {gl4_welch:.1f} → ✅ D")

# ─────────────────────────────────────────────────────────────────────────────
# 📊 GRÁFICA SECCIÓN 2
# ─────────────────────────────────────────────────────────────────────────────
fig2, axes2 = plt.subplots(2, 1, figsize=(12, 7))
fig2.suptitle(f'Sección 2 – IC para Diferencia de Medias (Muestras Pequeñas)\n{NOMBRE}',
              fontsize=13, fontweight='bold', color='#1A237E')

plot_ic_horizontal(axes2[0], li3, ls3, diff3, cero=0,
                   titulo='P5-P7 | IC 90%: Metodologías Educativas (μI − μT)',
                   etiqueta=f'IC 90%: ({li3:.2f}, {ls3:.2f})', color='#E91E63')

plot_ic_horizontal(axes2[1], li4, ls4, diff4, cero=0,
                   titulo='P8-P10 | IC 95%: Campaña Salud Pública (μA − μB)',
                   etiqueta=f'IC 95%: ({li4:.2f}, {ls4:.2f})', color='#009688')

plt.tight_layout()
plt.savefig('IC_Seccion2_MariaAhumadaLora.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Gráfica guardada: IC_Seccion2_MariaAhumadaLora.png")
```

```python
%%R
# =============================================================================
# @title R SECCIÓN 2: MUESTRAS PEQUEÑAS (t-Student)
# =============================================================================
cat("=================================================================\n")
cat("  SECCIÓN 2 – IC DIFERENCIA DE MEDIAS (MUESTRAS PEQUEÑAS)\n")
cat("=================================================================\n")

# Contexto 3
n_t <- 12; x_t <- 3.8; s_t <- 0.6
n_i <- 10; x_i <- 4.3; s_i <- 0.5
diff3 <- x_i - x_t
gl3   <- n_t + n_i - 2
sp2_3 <- ((n_t-1)*s_t^2 + (n_i-1)*s_i^2) / gl3
sp3   <- sqrt(sp2_3)
t90   <- qt(0.95, df = gl3)
se3   <- sp3 * sqrt(1/n_i + 1/n_t)
li3 <- diff3 - t90 * se3; ls3 <- diff3 + t90 * se3

cat(sprintf("\nP5-P7 | Metodologías -> t(%.0f)=%.3f | IC 90%%: (%.2f, %.2f)\n", gl3, t90, li3, ls3))
cat("  ✅ Respuesta: A, C, B\n\n")

# Contexto 4
n_a <- 15; x_a <- 12.5; s_a <- 2.8
n_b2 <- 13; x_b2 <- 10.2; s_b2 <- 3.0
diff4 <- x_a - x_b2
gl4   <- n_a + n_b2 - 2
sp2_4 <- ((n_a-1)*s_a^2 + (n_b2-1)*s_b2^2) / gl4
sp4   <- sqrt(sp2_4)
t95   <- qt(0.975, df = gl4)
se4   <- sp4 * sqrt(1/n_a + 1/n_b2)
li4 <- diff4 - t95*se4; ls4 <- diff4 + t95*se4

cat(sprintf("P8-P10 | Salud -> GL=%d | IC 95%%: (%.2f, %.2f)\n", gl4, li4, ls4))
cat("  ✅ Respuesta: A, C, B\n\n")

# ── GRÁFICAS SECCIÓN 2 ────────────────────────────────────────────────────────
g2a <- plot_ic(li3, ls3, diff3,
               titulo = "P5-P7 | IC 90% Metodologías Educativas",
               etiqueta = sprintf("IC: (%.2f, %.2f)", li3, ls3),
               color_ic = "#E91E63")

g2b <- plot_ic(li4, ls4, diff4,
               titulo = "P8-P10 | IC 95% Campaña Salud",
               etiqueta = "No contiene 0 -> Efectivo",
               color_ic = "#009688")

print(grid.arrange(g2a, g2b, nrow = 2, top = paste("Sección 2 |", NOMBRE)))

png("Seccion2_IC.png", width = 1200, height = 600, res = 150)
grid.arrange(g2a, g2b, nrow = 2, top = paste("Sección 2 |", NOMBRE))
dev.off()
cat("✅ Gráfica guardada: Seccion2_IC.png\n")
```

## **SECCIÓN 3: INTERVALOS DE CONFIANZA PARA DIFERENCIA DE PROPORCIONES**
Contexto 5: Intención de voto en dos ciudades
Una firma encuestadora quiere comparar la intención de voto por un candidato en dos ciudades principales. En Bogotá, de 500 encuestados, 275 apoyan al candidato. En Medellín, de 400 encuestados, 200 lo apoyan.

Pregunta 11. ¿Cuál es la proporción muestral en cada ciudad?

* p̂B = 0.52, p̂M = 0.48
* p̂B = 0.55, p̂M = 0.50
* p̂B = 0.58, p̂M = 0.52
* p̂B = 0.50, p̂M = 0.45

Pregunta 12. ¿Cuál es el error estándar para la diferencia de proporciones?

* 0.032
* 0.035
* 0.028
* 0.041

Pregunta 13. ¿Cuál es el intervalo de confianza del 95% para la verdadera diferencia de proporciones (pB - pM)?

* (0.05 ± 0.07)
* (0.03 ± 0.06)
* (0.07 ± 0.05)
* (0.05 ± 0.04)

Pregunta 14. Con base en el intervalo anterior, ¿se puede afirmar que hay diferencia significativa entre las dos ciudades?

* Sí, porque el intervalo no contiene el cero
* No, porque el intervalo contiene el cero
* Sí, porque las muestras son grandes
* No, porque las proporciones son muy cercanas


```python
# =============================================================================
# @title PYTHON SECCIÓN 3: DIFERENCIA DE PROPORCIONES PYTHON
# =============================================================================
import numpy as np
import matplotlib.pyplot as plt

# Define the plot_ic_horizontal function
def plot_ic_horizontal(ax, li, ls, est, cero=0, titulo='', etiqueta='', color='#2196F3'):
    ax.hlines(1, li, ls, colors=color, lw=5, label=etiqueta)
    ax.plot(est, 1, 'o', color='#FF9800', markersize=10, label='Estimación puntual')
    ax.axvline(cero, color='#F44336', linestyle='--', lw=1.5, label='Valor de referencia (0)')

    ax.set_yticks([])
    ax.set_ylim(0.8, 1.2)
    ax.set_title(titulo, fontsize=12, fontweight='bold', color='#1A237E')
    ax.text(li, 1.1, f'LI={li:.4f}', color=color, va='center', ha='right')
    ax.text(ls, 1.1, f'LS={ls:.4f}', color=color, va='center', ha='left')
    ax.text(est, 0.9, f'Est={est:.4f}', color='#FF9800', va='center', ha='center', fontsize=9)
    ax.grid(axis='x', linestyle='--', alpha=0.7)

NOMBRE = 'María Ahumada Lora'

print("\n" + "=" * 65)
print("  SECCIÓN 3: INTERVALOS DE CONFIANZA – PROPORCIONES")
print("=" * 65)

n_bog, x_bog = 500, 275
n_med, x_med = 400, 200
p_b5 = x_bog / n_bog
p_m5 = x_med / n_med
diff5 = p_b5 - p_m5
se5 = np.sqrt(p_b5*(1-p_b5)/n_bog + p_m5*(1-p_m5)/n_med)
margen5 = 1.960 * se5
li5 = diff5 - margen5
ls5 = diff5 + margen5

print(f"\n📊 PREGUNTA 11 – Proporciones: p̂B={p_b5:.2f}, p̂M={p_m5:.2f} → ✅ B")
print(f"\n📊 PREGUNTA 12 – Error estándar: SE={se5:.4f} → ✅ B")
print(f"\n📊 PREGUNTA 13 – IC 95%: ({li5:.4f} , {ls5:.4f}) → ✅ D")
print(f"\n📊 PREGUNTA 14 – ¿Diferencia significativa? IC contiene 0 → ✅ B")

# ─────────────────────────────────────────────────────────────────────────────
# 📊 GRÁFICA SECCIÓN 3
# ─────────────────────────────────────────────────────────────────────────────
fig3, ax3 = plt.subplots(1, 1, figsize=(12, 4))
fig3.suptitle(f'Sección 3 – IC 95% para Diferencia de Proporciones\n{NOMBRE}',
              fontsize=13, fontweight='bold', color='#1A237E')

plot_ic_horizontal(ax3, li5, ls5, diff5, cero=0,
                   titulo='P11-P14 | IC 95%: Intención de Voto Bogotá vs Medellín',
                   etiqueta=f'IC 95%: ({li5:.4f}, {ls5:.4f})', color='#FF5722')

plt.tight_layout()
plt.savefig('IC_Seccion3_MariaAhumadaLora.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Gráfica guardada: IC_Seccion3_MariaAhumadaLora.png")
```

```python
%%R
# =============================================================================
# @title R SECCIÓN 3: DIFERENCIA DE PROPORCIONES
# =============================================================================
cat("=================================================================\n")
cat("  SECCIÓN 3 – IC DIFERENCIA DE PROPORCIONES\n")
cat("=================================================================\n")

n_bog <- 500; x_bog <- 275
n_med <- 400; x_med <- 200
p_b5  <- x_bog / n_bog
p_m5  <- x_med / n_med
diff5 <- p_b5 - p_m5
se5   <- sqrt(p_b5*(1-p_b5)/n_bog + p_m5*(1-p_m5)/n_med)
li5 <- diff5 - qnorm(0.975) * se5
ls5 <- diff5 + qnorm(0.975) * se5

cat(sprintf("\nP11-P14 | Proporciones -> pB=%.2f, pM=%.2f\n", p_b5, p_m5))
cat(sprintf("IC 95%%: (%.4f, %.4f) -> Contiene 0 (No significativo)\n", li5, ls5))
cat("  ✅ Respuesta: B, A, A, B\n\n")

# ── GRÁFICA SECCIÓN 3 ─────────────────────────────────────────────────────────
g3 <- plot_ic(li5, ls5, diff5,
              titulo = "P11-P14 | IC 95% Proporciones (Bogotá vs Medellín)",
              etiqueta = "Contiene el 0 -> No hay diferencia significativa",
              color_ic = "#FF5722")

print(grid.arrange(g3, top = paste("Sección 3 |", NOMBRE)))

png("Seccion3_IC.png", width = 1200, height = 400, res = 150)
grid.arrange(g3, top = paste("Sección 3 |", NOMBRE))
dev.off()
cat("✅ Gráfica guardada: Seccion3_IC.png\n")
```

## **SECCIÓN 4: INTERVALOS DE CONFIANZA PARA LA VARIANZA**
Contexto 6: Control de calidad en producción agrícola
Un programa de certificación orgánica requiere que la varianza en el peso de los productos no supere cierto límite. Se tomó una muestra de 20 aguacates, obteniendo una varianza muestral s² = 25 gramos².

Pregunta 15. ¿Cuáles son los valores críticos χ² para un intervalo de confianza del 90% para la varianza poblacional?

* χ²0.05, 19 = 30.14, χ²0.95, 19 = 10.12
* χ²0.05, 20 = 31.41, χ²0.95, 20 = 10.85
* χ²0.05, 18 = 28.87, χ²0.95, 18 = 9.39
* χ²0.025, 19 = 32.85, χ²0.975, 19 = 8.91

Pregunta 16. ¿Cuál es el intervalo de confianza del 90% para la varianza poblacional?

* (15.8, 46.9)
* (18.2, 52.3)
* (16.4, 48.7)
* (17.5, 50.1)

Pregunta 17. Si el límite máximo permitido es 30 gramos², ¿qué concluye?

* La varianza está dentro del límite permitido
* No se puede concluir porque el intervalo contiene valores mayores a 30
* La producción cumple con la certificación
* La muestra es insuficiente para concluir


```python
# =============================================================================
# @title PYTHON SECCIÓN 4: INTERVALO DE CONFIANZA PARA LA VARIANZA (χ²)
# =============================================================================
print("\n" + "=" * 65)
print("  SECCIÓN 4: INTERVALOS DE CONFIANZA – VARIANZA")
print("=" * 65)

n6, s2_6 = 20, 25
gl6 = n6 - 1
chi2_alto = stats.chi2.ppf(0.95, df=gl6)
chi2_bajo = stats.chi2.ppf(0.05, df=gl6)
li6 = (gl6 * s2_6) / chi2_alto
ls6 = (gl6 * s2_6) / chi2_bajo

print(f"\n📊 PREGUNTA 15 – Valores críticos χ²")
print(f"   χ²(0.05,19)={chi2_alto:.2f}, χ²(0.95,19)={chi2_bajo:.2f} → ✅ A")
print(f"\n📊 PREGUNTA 16 – IC 90% varianza: ({li6:.1f} , {ls6:.1f}) → ✅ A")
print(f"\n📊 PREGUNTA 17 – ¿Varianza > 30? IC incluye valores >30 → ✅ B")

# ─────────────────────────────────────────────────────────────────────────────
# 📊 GRÁFICA SECCIÓN 4: Distribución χ²
# ─────────────────────────────────────────────────────────────────────────────
fig4, ax4 = plt.subplots(figsize=(12, 5))
fig4.suptitle(f'Sección 4 – IC 90% para Varianza Poblacional (χ²)\n{NOMBRE}',
              fontsize=13, fontweight='bold', color='#1A237E')

x_chi = np.linspace(0, 50, 500)
y_chi = stats.chi2.pdf(x_chi, df=gl6)

ax4.plot(x_chi, y_chi, color='#5C6BC0', lw=2.5, label=f'χ²(df={gl6})')
ax4.fill_between(x_chi, y_chi, where=(x_chi >= chi2_bajo) & (x_chi <= chi2_alto),
                 alpha=0.35, color='#42A5F5', label='Región 90%')
ax4.axvline(chi2_bajo, color='#EF5350', lw=2, ls='--', label=f'χ²(0.95)={chi2_bajo:.2f}')
ax4.axvline(chi2_alto, color='#EF5350', lw=2, ls=':', label=f'χ²(0.05)={chi2_alto:.2f}')

ax4.set_xlabel('Valores χ²', fontsize=11)
ax4.set_ylabel('Densidad', fontsize=11)
ax4.set_title(f'P15-P17 | IC varianza: ({li6:.1f}, {ls6:.1f}) g²', fontsize=11)
ax4.text(0.01, 0.03, f'Estudiante: {NOMBRE}', transform=ax4.transAxes,
         fontsize=8, color='gray', style='italic')
ax4.legend(fontsize=9)

plt.tight_layout()
plt.savefig('IC_Seccion4_MariaAhumadaLora.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Gráfica guardada: IC_Seccion4_MariaAhumadaLora.png")
```

```python
%%R
# =============================================================================
# @title R SECCIÓN 4: VARIANZA (χ²)
# =============================================================================
cat("=================================================================\n")
cat("  SECCIÓN 4 – IC PARA LA VARIANZA POBLACIONAL (χ²)\n")
cat("=================================================================\n")

n6  <- 20; s2_6 <- 25; gl6 <- n6 - 1
chi_hi <- qchisq(0.95, df = gl6)
chi_lo <- qchisq(0.05, df = gl6)
li6 <- gl6 * s2_6 / chi_hi
ls6 <- gl6 * s2_6 / chi_lo

cat(sprintf("\nP15-P17 | Varianza -> χ²_lo=%.2f, χ²_hi=%.2f\n", chi_lo, chi_hi))
cat(sprintf("IC 90%%: (%.1f , %.1f) -> Contiene 30 (No conclusivo)\n", li6, ls6))
cat("  ✅ Respuesta: A, A, B\n\n")

# ── GRÁFICA SECCIÓN 4 (Distribución) ──────────────────────────────────────────
x_c <- seq(0, 50, length.out = 500)
df_c <- data.frame(x = x_c, y = dchisq(x_c, df = gl6))
df_c$region <- ifelse(df_c$x < chi_lo | df_c$x > chi_hi, "Cola", "IC 90%")

g4 <- ggplot(df_c, aes(x, y)) +
  geom_area(aes(fill = region), alpha = 0.45) +
  geom_line(color = "#5C6BC0", linewidth = 1.5) +
  geom_vline(xintercept = c(chi_lo, chi_hi), color = "#EF5350", linetype = "dashed") +
  scale_fill_manual(values = c("Cola" = "#EF5350", "IC 90%" = "#42A5F5")) +
  annotate("text", x = 35, y = 0.06, label = sprintf("IC: (%.1f, %.1f)", li6, ls6),
           size = 4, color = "#1A237E", fontface = "bold") +
  labs(title = "P15-P17 | Distribución χ² - IC Varianza", x = "Valores", y = "Densidad") +
  theme_minimal(base_size = 11) +
  theme(plot.title = element_text(face = "bold", color = "#1A237E"))

print(grid.arrange(g4, top = paste("Sección 4 |", NOMBRE)))

png("Seccion4_IC.png", width = 1200, height = 500, res = 150)
grid.arrange(g4, top = paste("Sección 4 |", NOMBRE))
dev.off()
cat("✅ Gráfica guardada: Seccion4_IC.png\n")
```

## **SECCIÓN 5: INTERVALOS DE CONFIANZA PARA COCIENTE DE VARIANZAS**
Contexto 7: Comparación de volatilidad en mercados financieros
Un analista económico compara la volatilidad (varianza de rendimientos) entre dos sectores: tecnología y servicios públicos. Muestras: tecnología n₁ = 25, varianza s₁² = 0.042; servicios públicos n₂ = 20, varianza s₂² = 0.018.

Pregunta 18. ¿Cuál es el estadístico F (cociente de varianzas muestrales)?

* F = 2.33
* F = 2.00
* F = 2.50
* F = 1.80

Pregunta 19. ¿Cuáles son los grados de libertad para el intervalo?

* gl₁ = 24, gl₂ = 19
* gl₁ = 25, gl₂ = 20
* gl₁ = 24, gl₂ = 20
* gl₁ = 25, gl₂ = 19

Pregunta 20. ¿Cuál es el intervalo de confianza del 95% para el cociente de varianzas poblacionales (σ₁²/σ₂²)?

* (1.15, 4.82)
* (1.08, 4.51)
* (1.22, 5.10)
* (0.95, 4.20)




```python
# =============================================================================
# @title PYTHON SECCIÓN 5: COCIENTE DE VARIANZAS (distribución F)
# =============================================================================
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

NOMBRE = 'María Ahumada Lora'

print("\n" + "=" * 65)
print("  SECCIÓN 5: INTERVALOS DE CONFIANZA – COCIENTE DE VARIANZAS")
print("=" * 65)

n1, s2_1 = 25, 0.042
n2, s2_2 = 20, 0.018
F_obs = s2_1 / s2_2
gl1_f = n1 - 1
gl2_f = n2 - 1

F_alto = stats.f.ppf(0.975, dfn=gl1_f, dfd=gl2_f)
F_bajo = stats.f.ppf(0.025, dfn=gl1_f, dfd=gl2_f)
li7 = F_obs / F_alto
ls7 = F_obs * stats.f.ppf(0.975, dfn=gl2_f, dfd=gl1_f)

print(f"\n📊 PREGUNTA 18 – Estadístico F = {F_obs:.4f} → ✅ A")
print(f"\n📊 PREGUNTA 19 – GL: gl1={gl1_f}, gl2={gl2_f} → ✅ A")
print(f"\n📊 PREGUNTA 20 – IC 95% cociente varianzas: ({li7:.2f} , {ls7:.2f}) → ✅ D")

# ─────────────────────────────────────────────────────────────────────────────
# 📊 GRÁFICA SECCIÓN 5: Distribución F
# ─────────────────────────────────────────────────────────────────────────────
fig5, ax5 = plt.subplots(figsize=(12, 5))
fig5.suptitle(f'Sección 5 – IC 95% para Cociente de Varianzas (F)\n{NOMBRE}',
              fontsize=13, fontweight='bold', color='#1A237E')

x_f = np.linspace(0, 6, 500)
y_f = stats.f.pdf(x_f, dfn=gl1_f, dfd=gl2_f)
f_lo = stats.f.ppf(0.025, dfn=gl1_f, dfd=gl2_f)
f_hi = stats.f.ppf(0.975, dfn=gl1_f, dfd=gl2_f)

ax5.plot(x_f, y_f, color='#7B1FA2', lw=2.5, label=f'F({gl1_f},{gl2_f})')
ax5.fill_between(x_f, y_f, where=(x_f >= f_lo) & (x_f <= f_hi),
                 alpha=0.35, color='#CE93D8', label=f'Región 95%')
ax5.axvline(F_obs, color='#FF9800', lw=2.5, ls='-.', label=f'F observado = {F_obs:.2f}')
ax5.axvline(li7, color='#1565C0', lw=2, ls='--', label=f'LI = {li7:.2f}')
ax5.axvline(ls7, color='#1565C0', lw=2, ls=':', label=f'LS = {ls7:.2f}')

ax5.set_xlabel('Valores F', fontsize=11)
ax5.set_ylabel('Densidad', fontsize=11)
ax5.set_title(f'P18-P20 | IC 95%: ({li7:.2f}, {ls7:.2f})', fontsize=11)
ax5.text(0.01, 0.03, f'Estudiante: {NOMBRE}', transform=ax5.transAxes,
         fontsize=8, color='gray', style='italic')
ax5.legend(fontsize=9)

plt.tight_layout()
plt.savefig('IC_Seccion5_MariaAhumadaLora.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Gráfica guardada: IC_Seccion5_MariaAhumadaLora.png")
```

```python
%%R
# =============================================================================
# @title R SECCIÓN 5: COCIENTE DE VARIANZAS (F)
# =============================================================================
install.packages('gridExtra', repos='http://cran.us.r-project.org')
library(ggplot2)
library(gridExtra)

NOMBRE <- 'María Ahumada Lora'

cat("=================================================================\n")
cat("  SECCIÓN 5 – IC PARA COCIENTE DE VARIANZAS (F)\n")
cat("=================================================================\n")

n1 <- 25; s2_1 <- 0.042; gl1f <- n1 - 1
n2 <- 20; s2_2 <- 0.018; gl2f <- n2 - 1
F_obs <- s2_1 / s2_2
F_hi  <- qf(0.975, df1 = gl1f, df2 = gl2f)
F_lo  <- qf(0.025, df1 = gl1f, df2 = gl2f) # Límite inferior de la distribución F
li7 <- F_obs / F_hi
# Límite superior usando la propiedad F_1-alpha(dfn,dfd) = 1/F_alpha(dfd,dfn)
ls7 <- F_obs * qf(0.975, df1 = gl2f, df2 = gl1f)

cat(sprintf("\nP18-P20 | F-ratio -> F_obs=%.2f\n", F_obs))
cat(sprintf("IC 95%%: (%.2f , %.2f) -> No contiene 1 (Diferentes varianzas)\n", li7, ls7))
cat("  ✅ Respuesta: A, A, D\n\n")

# ── GRÁFICA SECCIÓN 5 (Distribución F) ────────────────────────────────────────
x_f  <- seq(0, 6, length.out = 500)
df_f <- data.frame(x = x_f, y = df(x_f, df1 = gl1f, df2 = gl2f))
df_f$region <- ifelse(df_f$x < F_lo | df_f$x > F_hi, "Cola", "IC 95%")

g5 <- ggplot(df_f, aes(x, y)) +
  geom_area(aes(fill = region), alpha = 0.45) +
  geom_line(color = "#7B1FA2", linewidth = 1.5) +
  geom_vline(xintercept = F_obs, color = "#FF9800", linetype = "dotdash", linewidth = 1.5) +
  geom_vline(xintercept = c(li7, ls7), color = "#1565C0", linetype = "dashed") +
  scale_fill_manual(values = c("Cola" = "#EF5350", "IC 95%" = "#CE93D8")) +
  annotate("text", x = 4.5, y = 0.6, label = sprintf("IC: (%.2f, %.2f)", li7, ls7),
           size = 4, color = "#1A237E", fontface = "bold") +
  labs(title = "P18-P20 | Distribución F - IC Cociente", x = "Valores", y = "Densidad") +
  theme_minimal(base_size = 11) +
  theme(plot.title = element_text(face = "bold", color = "#1A237E"))

print(grid.arrange(g5, top = paste("Sección 5 |", NOMBRE)))

png("Seccion5_IC.png", width = 1200, height = 500, res = 150)
grid.arrange(g5, top = paste("Sección 5 |", NOMBRE))
dev.off()
cat("✅ Gráfica guardada: Seccion5_IC.png\n")
```
